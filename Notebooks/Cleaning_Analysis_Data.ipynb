# 1.Import libraries and data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#categorical data
categorical=data[[ 'blue', 'dual_sim', 'four_g', 'wifi', 'price_range']]
```

```python
#numaric data
num=data[['battery_power', 'clock_speed','int_memory', 'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width','ram', 'sc_h', 'sc_w']]
```

# 3.Missing Values Handling

```python
for col  in categorical.columns:
  data[col].fillna(data[col].mode()[0],inplace=True)
```

```python
for col in num.columns:
  data[col].fillna(data[col].median(),inplace=True)
```

```python
for col in num.columns:
  data[col].replace(0,data[col].median(),inplace=True)
```

# 4.Duplicate Rows

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

# 5.Outliers Detection and Treatment

```python
#show the outliers
for col in num.columns:
  plt.figure(figsize=(10,10))
  plt.boxplot(data[col])
  plt.title(col)
  plt.show()
```

```python
q1=data['n_cores'].quantile(0.25)
q3=data['n_cores'].quantile(0.75)
iqr=q3-q1
upper=q3+1.5*iqr
lower=q1-1.5*iqr
data['n_cores']=np.where(data['n_cores']>upper,upper,data['n_cores'])
```

```python
q1=data['px_height'].quantile(0.25)
q3=data['px_height'].quantile(0.75)
iqr=q3-q1
upper=q3+1.5*iqr
lower=q1-1.5*iqr
data['px_height']=np.where(data['px_height']>upper,upper,data['px_height'])
```

```python
for col in num.columns:
  plt.figure(figsize=(10,10))
  plt.boxplot(data[col])
  plt.title(col)
  plt.show()
```

# 6.Feature Distribution Analysis

```python
plt.figure(figsize=(10,5))
plt.pie(data['price_range'].value_counts(),labels=data['price_range'].value_counts().index,autopct='%1.3f%%')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.pie(data['blue'].value_counts(),labels=data['blue'].value_counts().index,autopct='%1.2f%%')
plt.show()
```

**bar charts for other categorical columns**

```python
#categorical other blue and price range
goal=data[['dual_sim','four_g','wifi']]
```

**histogram for numeric columns**

```python
#for numeric columns
for col in num.columns:
  plt.figure(figsize=(10,5))
  plt.hist(data[col],bins=10,color='skyblue',edgecolor='black')
  plt.title(col)
  plt.xlabel(col)
  plt.ylabel('frequancy')
  plt.show()
```

**countplot for numeric columns**

```python
#for categorical columns
for col in categorical.columns:
  plt.figure(figsize=(8, 4))
  sns.countplot(data=data, x=col,color='darkblue',)
  plt.title(col)
  plt.show()
```
